```python
import pandas as pd

# Create the DataFrame
df = pd.DataFrame({
    'Name': ['Danyal Ahmed', 'Ali', 'Talha'],
    'Age': [21, 19, 17],
    'Gender': ['Male', 'Male', 'Male']
})

# Print only the 'Age' column
print(df['Age'])

# Save the entire DataFrame to an Excel file
df.to_excel('practice.xlsx', sheet_name='practice', index=False)
```
